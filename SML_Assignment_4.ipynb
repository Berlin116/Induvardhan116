```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dis=pd.read_csv("/content/train.csv")
print(dis)
```

```python
# find features and target variables
y=dis['price_range']
print(y)
print(y.shape)
```

```python
x=dis.drop('price_range',axis=1)
print(x)
```

```python
# Normalize the data with min-max scaling
xx = (x-x.min())/(x.max()-x.min())
print(xx)
```

```python
# Split the data into train and test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xx,y,test_size=0.2,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
#Question 2
dis=pd.read_csv("/content/housing.csv")
print(dis)
```

```python
# Find data type and shape
y=dis['ocean_proximity']
print(y)
print(y.shape)
print(y.dtype)
```

```python
# Find the target and features
x = dis.drop('ocean_proximity',axis=1)
print(x)
```

```python
# Normalize all the features
xx = (x-x.min())/(x.max()-x.min())
print(xx)
```

```python
# Split the data into train and test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xx,y,test_size=0.2,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
